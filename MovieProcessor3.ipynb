##Extract features of 2018 movies from wikipedia


```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
movie2018 = df[['Title','Cast and crew','genres']]
```

```python
def get_director_name(x):
  if " (director)" in x:
    return x.split(" (director)")[0]
  elif " (directors)" in x:
    return x.split(" (directors)")[0]
  else:
    return x.split(" (director/screenplay)")[0]
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python

def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
movie2018['director_name'] = movie2018['Cast and crew'].map(lambda x: get_director_name(x))
```

```python
movie2018['actor_1_name'] = movie2018['Cast and crew'].map(lambda x: get_actor1(x))
movie2018['actor_2_name'] = movie2018['Cast and crew'].map(lambda x: get_actor2(x))
movie2018['actor_3_name'] = movie2018['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
movie2018 = movie2018.rename(columns={'Title':'movie_title'})
movie2018 = movie2018.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
movie2018.isnull().sum()
```

```python
movie2018['actor_2_name'] = movie2018['actor_2_name'].replace(np.nan, 'unknown')
movie2018['actor_3_name'] = movie2018['actor_3_name'].replace(np.nan, 'unknown')
movie2018['movie_title'] = movie2018['movie_title'].str.lower()
movie2018['comb'] = movie2018['actor_1_name'] + ' ' + movie2018['actor_2_name'] + ' '+ movie2018['actor_3_name'] + ' '+ movie2018['director_name'] +' ' + movie2018['genres']
```

```python
movie2018.dropna(inplace=True)
```
