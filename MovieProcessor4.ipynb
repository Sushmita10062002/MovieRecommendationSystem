```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

```python
df.drop(axis=0,index=127,inplace=True)
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
movie2020 = df[['Title','Cast and crew','genres']]
```

```python

def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]

def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
movie2020['director_name'] = movie2020['Cast and crew'].map(lambda x: get_director(x))
movie2020['actor_1_name'] = movie2020['Cast and crew'].map(lambda x: get_actor1(x))
movie2020['actor_2_name'] = movie2020['Cast and crew'].map(lambda x: get_actor2(x))
movie2020['actor_3_name'] = movie2020['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
movie2020 = movie2020.rename(columns={'Title':'movie_title'})
movie2020 = movie2020.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
movie2020['actor_2_name'] = movie2020['actor_2_name'].replace(np.nan, 'unknown')
movie2020['actor_3_name'] = movie2020['actor_3_name'].replace(np.nan, 'unknown')
movie2020['movie_title'] = movie2020['movie_title'].str.lower()
movie2020['comb'] = movie2020['actor_1_name'] + ' ' + movie2020['actor_2_name'] + ' '+ movie2020['actor_3_name'] + ' '+ movie2020['director_name'] +' ' + movie2020['genres']
```
